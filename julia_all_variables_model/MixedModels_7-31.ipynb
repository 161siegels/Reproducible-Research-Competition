```julia
using CSV, DataFrames, MixedModels, StatsModels, RCall
```

```julia
train2 = CSV.read("scaled_train.csv")
```

```julia
form = @formula(SHOT_RESULT2 ~ FG_Percent+FT_Percent+FG3_Percent+layup+dunk+hook+fadeaway+pullup+running+driving+reverse+fingerroll+putback+turnaround+bank+point_diff+tip+TOUCH_TIME+DRIBBLES+ratio+jump_shot+End_SC+End_PC+LOCATION+Quick_S+Clutch_T+Garbage_T+SHOT_CLOCK+SHOT_NUMBER+(FG_Percent+FT_Percent+FG3_Percent+layup+dunk+hook+fadeaway+pullup+running+driving+reverse+fingerroll+putback+turnaround+bank+point_diff+tip+TOUCH_TIME+DRIBBLES+ratio+jump_shot+End_SC+End_PC+LOCATION+Quick_S+Clutch_T+Garbage_T+SHOT_CLOCK+SHOT_NUMBER+1 | player_name))
```

```julia
md1 = GeneralizedLinearMixedModel(form, train2, Bernoulli())
```

```julia
glm1 = @time pirls!(md1, true, true)
```

```julia
r_effects = ranef(glm1, named = true)
r_effects[1]
```

```julia
r_effects = ranef(glm1, named = true)
player_order = names(r_effects[1])[2]
r_effects2 = convert(Array, r_effects[1])

@rput(r_effects2)
@rput(player_order)
R"write.csv(file = 'random_effects.csv', r_effects2)"
R"write.csv(file = 'player_names.csv', player_order)"
# # I continued this in Rmd because it was easier to work with, that's why we're on random_effects2

# # create fixed effects csv
fixed = coef(glm1)
@rput(fixed)
R"write.csv(file = 'fixed_effects.csv', fixed)"
```
